```python
import pandas as pd

# Load the datasets
elliptic_txs_edgelist = pd.read_csv('elliptic_bitcoin_dataset\elliptic_txs_edgelist.csv')
elliptic_txs_classes = pd.read_csv('elliptic_bitcoin_dataset\elliptic_txs_classes.csv')

# Display the first few rows of each dataset
elliptic_txs_edgelist.head(), elliptic_txs_classes.head()

```

```python
import matplotlib.pyplot as plt

# Check the distribution of the classes
class_distribution = elliptic_txs_classes['class'].value_counts()

# Plotting the distribution
plt.figure(figsize=(10, 6))
class_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Classes in elliptic_txs_classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import networkx as nx

# Create a graph from the edgelist
G = nx.from_pandas_edgelist(elliptic_txs_edgelist.head(1000), 'txId1', 'txId2')

# Plotting the graph
plt.figure(figsize=(12, 12))
nx.draw(G, with_labels=False, node_size=10, node_color='skyblue', edge_color='gray')
plt.title('Network of Transactions (Subset)')
plt.show()
```
